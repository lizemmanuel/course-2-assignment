### Service settings:

```python
# Plot the appointments over the available date range, and review the service settings for months.
# Create a lineplot.
```

```python

nc_ss = nc.groupby(['Apointment_month','Service_setting'])['Count_of_appointments']\
.agg(sum).reset_index()

print(nc_ss)
```

```python

nc_ct = nc.groupby(['Apointment_month','Context_type'])['Count_of_appointments']\
.agg(sum).reset_index()

print(nc_ct)
```

```python

# Create empty plot and set plot size. 
sns.set(rc={'figure.figsize':(15,12)})

# Set the plot style as white.
sns.set_style('white')

# Create lineplot for service settings
sns.lineplot(x='Apointment_month', y='Count_of_appointments',\
             hue='Service_setting', ci=None, data=nc_ss)
```
